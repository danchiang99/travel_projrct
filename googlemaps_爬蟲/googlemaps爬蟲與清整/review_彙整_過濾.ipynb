```python
import os
import pandas as pd

def read_all_csv_files_in_folder(folder_path):
    # 获取指定文件夹内的所有文件名
    file_list = os.listdir(folder_path)

    # 创建一个空的DataFrame，用于存储合并后的数据
    combined_df = pd.DataFrame()

    # 循环遍历文件列表，读取所有CSV文件的内容，并将其合并到combined_df中
    
    for filename in file_list:
        if filename.lower().endswith('.csv'):  # 检查文件扩展名是否为CSV
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            combined_df = pd.concat([combined_df, df], ignore_index=True)

    return combined_df
```

```python
import time
# 示例文件夹路径
folder_path = 'review/'
start_time = time.time() 
# 读取文件夹内所有CSV文件的内容并合并为一个DataFrame
all_df = read_all_csv_files_in_folder(folder_path)
all_df = all_df.drop(['Unnamed: 0'],axis = 1)
end_time = time.time() 
print("run times: ",int((end_time-start_time)/3600),"hr",int((end_time-start_time)/60)-int((end_time-start_time)/3600)*60,"min",round((end_time-start_time)%60,1),"sec")
all_df
```

```python
# 排除多餘欄位
result_df = all_df[['rw_name','rw_rating_total','rw_rating_count_total','rw_rating_reviewer','rw_review']]
result_df
```

```python
result_df = result_df[result_df['rw_review'].apply(lambda x: len(str(x)) >= 30)]
result_df = result_df.reset_index(drop=True)
result_df
```

```python
result_df.to_csv('review_result_len_more30.csv',encoding='utf-8-sig')
```

```python
result_df = result_df[result_df['rw_review'].apply(lambda x: len(str(x)) <= 150)]
result_df = result_df.reset_index(drop=True)
result_df
```

```python
# 隨機打散
result_df = result_df.sample(frac=1)  # frac=1 表示抽取全部數據，相當於完全打散
```

```python
# 根據評分排序
result_df = result_df.sort_values(by='rw_rating_reviewer', ascending=True)
result_df
```

```python
result_df.to_csv('review_result_random_sort_start.csv',encoding='utf-8-sig')
```

```python
# 定义一个函数来获取每个分组的前 1000 行数据
def get_top_n(group):
    return group.head(2000)

data = pd.read_csv('review_result_random_sort_start.csv')
data = data.drop(['Unnamed: 0'],axis = 1)

# 根据 'Category' 列分组，并应用上面定义的函数
random_get1000_result_df = data.groupby('rw_rating_reviewer', group_keys=False).apply(get_top_n)
# random_get1000_result_df.reset_index(drop=True)
random_get1000_result_df
```

```python
random_get1000_result_df.to_csv('random_get1000_groupbystart_result.csv',encoding='utf-8-sig')
```
