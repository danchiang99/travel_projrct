```python
import os
import pandas as pd

# 指定包含CSV文件的文件夹路径
folder_path = 'review/'

# 列出文件夹中的所有CSV文件
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# 针对每个CSV文件进行修正
for file in csv_files:
    file_path = os.path.join(folder_path, file)

    # 读取CSV文件为DataFrame
    data = pd.read_csv(file_path)

    # 判断是否存在 'rw_iongitude' 列，如果存在则替换为 'rw_longitude'
    if 'rw_iongitude' in data.columns:
        data.rename(columns={'rw_iongitude': 'rw_longitude'}, inplace=True)

        # 将修正后的数据保存回CSV文件
        data.to_csv(file_path, index=False,encoding='utf-8-sig')

        print(f"修正了文件：{file}")

print("所有文件修正完成。")
```

```python
import os
import pandas as pd

# 指定包含CSV文件的文件夹路径
folder_path = 'search/'

# 列出文件夹中的所有CSV文件
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# 针对每个CSV文件进行修正
for file in csv_files:
    file_path = os.path.join(folder_path, file)

    # 读取CSV文件为DataFrame
    data = pd.read_csv(file_path)

    # 判断是否存在 'rw_iongitude' 列，如果存在则替换为 'rw_longitude'
    if 'iongitude' in data.columns:
        data.rename(columns={'iongitude': 'longitude'}, inplace=True)

        # 将修正后的数据保存回CSV文件
        data.to_csv(file_path, index=False,encoding='utf-8-sig')

        print(f"修正了文件：{file}")

print("所有文件修正完成。")
```

```python
import os
import pandas as pd

# 指定包含CSV文件的文件夹路径
folder_path = 'review/'

# 列出文件夹中的所有CSV文件
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# 针对每个CSV文件进行修正
for file in csv_files:
    file_path = os.path.join(folder_path, file)

    # 读取CSV文件为DataFrame
    data = pd.read_csv(file_path)

    # 判断是否存在 'rw_iongitude' 列，如果存在则替换为 'rw_longitude'
    if 'w_introduction' in data.columns:
        data.rename(columns={'w_introduction': 'rw_introduction'}, inplace=True)

        # 将修正后的数据保存回CSV文件
        data.to_csv(file_path, index=False,encoding='utf-8-sig')

        print(f"修正了文件：{file}")

print("所有文件修正完成。")
```
