```python
 # 記錄開始時間
start_time = time.time() 

#---- driver 設定 ----
s = Service("./geckodriver.exe")
# s = Service("./chromedriver.exe")
driver = webdriver.Chrome(service=s)
# 最小視窗
# driver.minimize_window()
# 最大視窗
driver.maximize_window() 
url ='https://www.google.com/maps'
driver.get(url)
timesleep_rand()

# 輸入查詢關鍵字
search_keyword = '臺北市 景點推薦'
search_box = driver.find_element(By.NAME, 'q')
search_box.send_keys(search_keyword)
timesleep_rand()
# 送出關鍵字
btn_search = driver.find_element(By.CLASS_NAME, "mL3xi")
btn_search.click()

# 相關查詢參數輸入
search_max = 100 #設定要搜尋多少筆資料(如果google有的話) 最小設定11
sort_type = 1 #0 : 最相關 1: 最新 2: 評分最高 3: 評分最低 
setdefault_max_review_slip =  50# google評論爬蟲查詢至1130筆後會沒有資料 故建議設定最多滑100次(滑1次10筆資料) 如果程式出錯大概率錯這因google好像會阻擋 錯的話就挑低此參數值 
search_default = 11 # 不可調整參數
refresh_flag = True # T 開啟 F 關閉


#---- 變數初始化 ----
# 搜尋結果
io_current_time_list = []
io_keyword_list = []
io_name_list = []
io_rating_list = []
io_rating_count_list = []
io_category_list = []
io_address_list = []
io_introduce_list = []
io_Website_list = []
io_img_list = []
# 評論
rw_open_time_list = []
rw_category_list = []
rw_tag_list=[]
rw_current_time_list = []
rw_name_list = []
rw_latitude_list = []
rw_iongitude_list = []
rw_introduction_list = []
rw_rating_total_list = []
rw_rating_count_total_list = []
rw_phone_list = []
rw_address_list = []
rw_reviewer_list = []
rw_reviewer_page_list = []
rw_reviewer_tag_list = []
rw_review_time_list = []
rw_rating_reviewer_list = []
rw_likes_list = []
rw_review_list = []
rw_review_word_list = []
# 評論參數
rw_opentime = ""
rw_category = ""
rw_rating_total = ""
rw_rating_count_total = ""
rw_latitude = ""
rw_iongitude = ""
# 時間
time_refresh = 0

# 搜尋結果往下捲動
# 第一批搜尋結果的XPATH從3開始到11，每一個結果號碼+2，第一次迭代直接拉到最下方，後續每次移動1個結果，就可以保證一直產生新的結果
timesleep_rand()
tmpadd = 2
for i in range(search_default, search_max, tmpadd):
    search_default += tmpadd
    try:
        X_btn = driver.find_element(By.CLASS_NAME, 'VfPpkd-icon-LgbsSe')
        X_btn.click()
    except:
        time.sleep(1)

    try:        

        scrollable_search_result_block = driver.find_element(By.XPATH, f'//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]/div[{i}]/div/a')
    except:
        break
    else:
        driver.execute_script("arguments[0].scrollIntoView();", scrollable_search_result_block)
        timesleep_rand()


#---- 查詢html ----
timesleep_rand()
introduction = BeautifulSoup(driver.page_source, "html.parser")
total_result = introduction.find_all('a',class_='hfpxzc')
timesleep_rand()

#---- get 搜尋結果 ----
get_io_result()    
io_df = pd.DataFrame({
                        'keyword':io_keyword_list,
                        'name':io_name_list,
                        'current_time':io_current_time_list,
                        'rating':io_rating_list,
                        'rating_count':io_rating_count_list,
                        'category':io_category_list,
                        'address':io_address_list,
                        'introduce':io_introduce_list,
                        'Website':io_Website_list,
                        'img':io_img_list
})
# print(io_df)

#---- 搜尋結果存至csv檔 ----
tmp_search_keyword = ''
for j in search_keyword:
    if search_keyword ==" ":
        tmp_search_keyword+="_"
    else:
        tmp_search_keyword+=j
io_df.to_csv('v4_use_firfox_search_'+tmp_search_keyword+'result.csv',encoding='utf-8-sig')
# io_df.to_csv('v3_use_chrom_search_'+tmp_search_keyword+'result.csv',encoding='utf-8-sig')
search_review_btn = driver.find_elements(By.CLASS_NAME, 'hfpxzc')
#---- 評論爬取 ----
for i in range(len(total_result)):
    timesleep_rand()
    search_click_max_try = 10
    while search_click_max_try>0:
        search_click_max_try -= 1
        try:
            # 依照順序點擊搜尋結果
            search_review_btn[i].click()
            search_click_max_try = 0
        except:
            if search_click_max_try == 0:
                # 不明原因進到這 就截圖看畫面想看要怎麼處理
                driver.save_screenshot('bug_img/'+str(i+1)+"error_notfound"+'try_'+'.png')
                clsoe_above_main_windows() # 關閉多於視窗
                print(i+1,'error not found',end=" ")
                print(io_name_list[i])
        else:
            timesleep_rand()
            
                
    timesleep_rand()
    #---- 獲取景點經緯度 ----
    try:
        rw_latitude = total_result[i]['href'].split("data=")[1].split("!")[5].split("d")[1]
    except:
        print(i+1,'error latitude',end=" ")
    try:
        rw_iongitude = total_result[i]['href'].split("data=")[1].split("!")[6].split("d")[1]
    except:
        print(i+1,'error iongitude',end=" ")

    #---- 查詢總攬html ----
    try:
        overview_review_result = BeautifulSoup(driver.page_source, "html.parser")
    except:
        print(i+1,'error overview_review_result')

    #---- 前往簡介 ----
    try:
        # 確認簡介是否在2的位置 是則執行內容
        if overview_review_result.find_all('div',class_='Gpq6kf')[2].text == "簡介":
            introduction_btn = driver.find_elements(By.CLASS_NAME, 'hh2c6')[2]
            introduction_btn.click()
            timesleep_rand()
            #---- 查詢簡介html ----
            introduction_review_result = BeautifulSoup(driver.page_source, "html.parser")
            #---- 返回 ----
            introduction_btn = driver.find_elements(By.CLASS_NAME, 'hh2c6')[0]
            introduction_btn.click()
            timesleep_rand()
        # 確認簡介是否在1的位置 是則執行內容
        elif overview_review_result.find_all('div',class_='Gpq6kf')[1].text == "簡介":
            introduction_btn = driver.find_elements(By.CLASS_NAME, 'hh2c6')[1]
            introduction_btn.click()
            timesleep_rand()
            #---- 查詢簡介html ----
            introduction_review_result = BeautifulSoup(driver.page_source, "html.parser")
            #---- 返回 ----
            introduction_btn = driver.find_elements(By.CLASS_NAME, 'hh2c6')[0]
            introduction_btn.click()
            timesleep_rand()
    except:
        clsoe_above_main_windows()
        print(i+1,' error - 前往簡介')
        
    #---- 營業時間處理 ----
    op_time_try_max = 3
    while op_time_try_max>0:
        op_time_try_max -=1
        try:
            # 營業有兩種不同模式，先確認是否是不需要切換頁面的模式
            rw_opentime_result = overview_review_result.find_all('tr',class_='y0skZc')
            if rw_opentime_result != []:
                tmp_1=""
                for ii in rw_opentime_result:
                    tmp_1 += ii.text
                    tmp_1 += " "
                rw_opentime = tmp_1
                op_time_try_max = 0
            # 否，則當成是點擊模式
            else:
                #---- 前往營業時間 ----
                introduction_btn = driver.find_elements(By.CLASS_NAME, class_='CsEnBe')[1]
                introduction_btn.click()
                timesleep_rand()
                #---- 查詢營業時間html ----
                opentime_result = BeautifulSoup(driver.page_source, "html.parser")
                #---- 返回 ----
                back_introduction_btn = driver.find_element(By.CLASS_NAME, 'VfPpkd-icon-LgbsSe')
                back_introduction_btn.click()
                timesleep_rand()
                tmp_2=opentime_result.find('div',class_='t39EBf')['aria-label']
                rw_opentime = tmp_2
                op_time_try_max = 0
            # 如果發生營業時間不存在 暫不處理 讓他進回圈找不到就出去
        except:
            # 嘗試搜尋結果點擊
            try:
                btn = driver.find_elements(By.CLASS_NAME, 'hfpxzc')[i]
                btn.click()
                timesleep_rand()
            except:
                 # 儲存照片做debug用
                driver.save_screenshot('bug_img/'+str(i+1)+"error_營業時間"+'try_'+str(3-op_time_try_max)+'.png')
                check_windows_num()
                clsoe_above_main_windows()
                print(i+1,' error - 營業時間' )
            

    # 分類
    rw_category=io_category_list[i]

    # 總評分與總評分人數
    rw_rating_total = io_rating_list[i]
    rw_rating_count_total = io_rating_count_list[i]
    timesleep_rand()
    
    #---- 前往評論 ----
    try:
        if overview_review_result.find_all('div',class_='Gpq6kf')[1].text == "評論":
            review_btn = driver.find_elements(By.CLASS_NAME, 'hh2c6')[1]
            review_btn.click()
            timesleep_rand()
    except:
        clsoe_above_main_windows()
        try:
            btn = driver.find_elements(By.CLASS_NAME, 'hfpxzc')[i]
            btn.click()
        except:
            print(review_btn)
            timesleep_rand()

    try:
        # 點選評論排序
        btn_sort_review = driver.find_element(By.CSS_SELECTOR, '#QA0Szd > div > div > div.w6VYqd > div.bJzME.Hu9e2e.tTVLSc > div > div.e07Vkf.kA9KIf > div > div > div.m6QErb.DxyBCb.kA9KIf.dS8AEf > div.m6QErb.Pf6ghf.KoSBEe.ecceSd.tLjsW > div.TrU0dc.kdfrQc > button')
        btn_sort_review.click()
        timesleep_rand()
        # 選取排序方式
        btn_sort_chose = driver.find_elements(By.CLASS_NAME, 'fxNQSd')[sort_type] 
        btn_sort_chose.click()
        timesleep_rand()

        # 評論往下捲動
        # 定位評論區塊
        scrollable_review_block = driver.find_element(By.CSS_SELECTOR, '#QA0Szd > div > div > div.w6VYqd > div.bJzME.Hu9e2e.tTVLSc > div > div.e07Vkf.kA9KIf > div > div > div.m6QErb.DxyBCb.kA9KIf.dS8AEf')
        # 因基礎10筆評論資料 每滑動1次多10筆 且執行到約1130筆後就google就爬不動(如下) 故最多只撈取前1000筆評論資料
        # review 20
        # review 30
        # ...
        # review 1130
        # review 1130
        slip_max = min(int((int(io_rating_count_list[i])-10)/10)+1,setdefault_max_review_slip)
        for tmp_slip_rw in range(slip_max):
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', scrollable_review_block)
            timesleep_rand()
    except:
        clsoe_above_main_windows()
        print(i+1,' error - 評論點擊or下拉' )
        
     #---- 點擊評論全文 ----
    try:
        all_word_btn = driver.find_elements(By.CLASS_NAME,'w8nwRe')
        for tmpclick in all_word_btn:
            if tmpclick.text == "全文":
                continue
            try :
                tmpclick.click()
                timesleep_rand()
            except:
                clsoe_above_main_windows()
                print(i+1,"error - 全文點擊失敗")
    except:
        clsoe_above_main_windows()
        print(i+1,' error - 全文點擊搜尋失敗' )
    try:
        review_result = BeautifulSoup(driver.page_source, "html.parser")
        review = review_result.find_all('div',class_='jJc9Ad')
    except:
        print(i+1,' error - 讀取所有留言失敗(大概率是因為跳出網頁後關閉造成)' )

    #---- get 評論結果 ----
    get_rw_result()
    
    rw_df = pd.DataFrame({
                            'rw_name':rw_name_list,
                            'rw_current_time':rw_current_time_list,
                            'rw_category':rw_category_list,
                            'rw_tag':rw_tag_list,
                            'rw_open_time':rw_open_time_list,
                            'rw_latitude':rw_latitude_list,
                            'rw_iongitude':rw_iongitude_list,
                            'w_introduction':rw_introduction_list,
                            'rw_rating_total':rw_rating_total_list,
                            'rw_rating_count_total':rw_rating_count_total_list,
                            'rw_phone':rw_phone_list,
                            'rw_address':rw_address_list,
                            'rw_reviewer':rw_reviewer_list,
                            'rw_reviewer_page':rw_reviewer_page_list,
                            'rw_reviewer_tag':rw_reviewer_tag_list,
                            'rw_review_time':rw_review_time_list,
                            'rw_rating_reviewer':rw_rating_reviewer_list,
                            'rw_likes':rw_likes_list,
                            'rw_review':rw_review_list,
                            'rw_review_word':rw_review_word_list
                        })
    end_time = time.time()  # 記錄結束時間
    # 每30分鐘重整一次
    execution_time_per = (end_time - start_time)/60/30
    
    if execution_time_per> time_refresh and refresh_flag == True:
        driver.refresh()
        timesleep_rand()
        time_refresh+=1
        tmpadd = 2
        for i in range(search_default, search_max, tmpadd):
            search_default += tmpadd
            try:
                X_btn = driver.find_element(By.CLASS_NAME, 'VfPpkd-icon-LgbsSe')
                X_btn.click()
            except:
                time.sleep(1)

            try:        

                scrollable_search_result_block = driver.find_element(By.XPATH, f'//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]/div[{i}]/div/a')
            except:
                break
            else:
                driver.execute_script("arguments[0].scrollIntoView();", scrollable_search_result_block)
                timesleep_rand()
    
    #---- 評論結果存至csv檔 ----
    rw_df.to_csv('v4_use_firfox_review_'+tmp_search_keyword+'result.csv',encoding='utf-8-sig')
#     rw_df.to_csv('v3_use_chrom_review_'+tmp_search_keyword+'result.csv',encoding='utf-8-sig')
    timesleep_rand() 

#---- 關閉瀏覽器 ----
# driver.close()
```
