```python
def read_all_csv_files_in_folder(folder_path):
    # è·å–æŒ‡å®šæ–‡ä»¶å¤¹å†…çš„æ‰€æœ‰æ–‡ä»¶å
    file_list = os.listdir(folder_path)

    # åˆ›å»ºä¸€ä¸ªç©ºçš„DataFrameï¼Œç”¨äºå­˜å‚¨åˆå¹¶åçš„æ•°æ®
    combined_df = pd.DataFrame()

    # å¾ªç¯éå†æ–‡ä»¶åˆ—è¡¨ï¼Œè¯»å–æ‰€æœ‰CSVæ–‡ä»¶çš„å†…å®¹ï¼Œå¹¶å°†å…¶åˆå¹¶åˆ°combined_dfä¸­
    for filename in file_list:
        if filename.lower().endswith('.csv'):  # æ£€æŸ¥æ–‡ä»¶æ‰©å±•åæ˜¯å¦ä¸ºCSV
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            combined_df = pd.concat([combined_df, df], ignore_index=True)

    return combined_df
```

```python
def fill_value_in_csv_files(folder_path, filename_to_modify, fill_value, target_column):
    # çµ„åˆç›®æ¨™æª”æ¡ˆçš„å®Œæ•´è·¯å¾‘
    file_path = os.path.join(folder_path, filename_to_modify)
    
    # è®€å– CSV æª”æ¡ˆ
    df = pd.read_csv(file_path)
    
    # å°‡æŒ‡å®šå€¼å¡«å…¥ç›®æ¨™æ¬„ä½
    df[target_column] = fill_value
    
    # å„²å­˜ä¿®æ”¹å¾Œçš„è³‡æ–™å›åˆ°åŸæª”æ¡ˆ
    df.to_csv(file_path, index=False, encoding='utf-8-sig')
```

```python
def check_file_existence(folder_path, file_name):
    # æ‹¼æ¥æ–‡ä»¶çš„å®Œæ•´è·¯å¾„
    file_path = os.path.join(folder_path, file_name)

    # ä½¿ç”¨ os æ¨¡å—çš„ isfile å‡½æ•°æ£€æŸ¥æ–‡ä»¶æ˜¯å¦å­˜åœ¨
    if os.path.isfile(file_path):
        return True
    else:
        return False
```
