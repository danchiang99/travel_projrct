```python
%matplotlib inline
from wordcloud import WordCloud#, STOPWORDS 
import matplotlib.pyplot as plt 
import jieba

city_worldcloud='新北市'
jieba.set_dictionary('chinexe_font/dict.txt')
# wordlist = list(jieba.cut(str(df_search_clear[df_search_clear['rw_city']=='新北市'].name)))
wordlist = list(jieba.cut(str(df_search_clear[df_search_clear['rw_city'] == city_worldcloud]['name'].values)))
words = " ".join(wordlist)

font = 'chinexe_font/SourceHanSansTW-Regular.otf'
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                # stopwords = stopwords, 
                min_font_size = 10,font_path=font).generate(words) 
  
# plot the WordCloud image      
# plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']              
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

print('新北市文字雲')
plt.show() 
wordcloud.to_file('img/'+city_worldcloud+'_word_cloud.png')
```
